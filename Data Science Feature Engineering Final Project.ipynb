```python
# File location and type
file_location = "/FileStore/tables/Melbourne_housing_FULL_1.csv"
file_type = "csv"

# CSV options
infer_schema = "false"
first_row_is_header = "true"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
df = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

display(df.head(10))
```

```python
#remove the undesirable columns
df = df.drop('Suburb','Address','Propertycount','SellerG','YearBuilt')
display(df.head(10))
```

```python
#find correlation between Landsize and Building area
from pyspark.ml.linalg import Vectors
from pyspark.ml.stat import Correlation
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
pd1=df2.toPandas()

col1 = pd1[~pd1['Landsize'].isna()][~pd1['BuildingArea'].isna()]['Landsize']
col2 = pd1[~pd1['Landsize'].isna()][~pd1['BuildingArea'].isna()]['BuildingArea']               


correlation = pd.to_numeric(col1).corr(pd.to_numeric(col2))

plt.xlabel("Landsize")
plt.ylabel("BuildingArea")
plt.title("Relationship between Landsize and BuildingArea")
plt.plot(pd.to_numeric(col1),pd.to_numeric(col2))

```

```python
#with percentile 0.01 and 0.99, remove outliers in 'distance'
pd2['Distance']=pd.to_numeric(pd2['Distance'])
unique_area = pd2['CouncilArea'].unique()
q1 = []
q99 = []
for each_unique_area in unique_area:
  each_unique_df = pd2[pd2['CouncilArea'] == each_unique_area]
  each_q1 = np.quantile(each_unique_df['Distance'], 0.01)
  each_q99 = np.quantile(each_unique_df['Distance'], 0.99)
  q1.append(each_q1)
  q99.append(each_q99)
quantile_df = pd.DataFrame()
quantile_df['CouncilArea'] = unique_area
quantile_df['q1'] = q1
quantile_df['q99'] = q99
merged_df = pd2.merge(quantile_df, how='left')
result_df = merged_df[(merged_df['Distance'] > merged_df['q1'])&(merged_df['Distance'] < merged_df['q99'])]
display(result_df)

pd3 = result_df
```

```python
from sklearn.preprocessing import StandardScaler

X_std = StandardScaler().fit_transform(pd3[["Price","Distance","Landsize"]])
pd5 = pd.DataFrame(X_std,columns = ["Price","Distance","Landsize"])
display(pd5)
```

```python
pd6 = pd3[[0,1,2,3,4,5,6,7]]
result = pd.concat([pd6,pd5], axis=1, sort=False)
display(result.head(5))
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

ks = range (1,15)
inertias =[]
for k in ks:
  model = KMeans(n_clusters=k)
  model.fit(result)
  inertias.append(model.inertia_)
plt.plot(ks,inertias,"-o",color = "blue")
plt.xlabel("Number of Clusters")
plt.ylabel("Loss Function")
plt.xticks(ks)
plt.show()
```

```python
model_final = KMeans(n_clusters=4)
labels = []
model_final.fit(result)
labels = model_final.labels_

result1 = pd.DataFrame()
result1["Price"] = pd3["Price"]
result1["Distance"] = pd3["Distance"]
result1["Landsize"] = pd3["Landsize"]
result1["Labels"] = labels
display(result1)
```
